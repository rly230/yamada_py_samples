フィード(RSS) の全 item 要素 の description 要素の内容を Sudachi で形態素解析

```python
# 必要なライブラリをインストール（Jupyter Notebookで実行する場合）
# !pip install beautifulsoup4 requests sudachipy

# 辞書をダウンロード（core）
# !pip install sudachidict_core

from sudachipy import dictionary
from sudachipy import tokenizer

tokenizer_obj = dictionary.Dictionary().create()
```

```python
import requests
from bs4 import BeautifulSoup

urlString = "https://rss.itmedia.co.jp/rss/2.0/itmedia_all.xml"
response = requests.get(urlString)

if response.status_code == 200:
    print("ページの取得に成功しました")
else:
    print(f"ページの取得に失敗しました: {response.status_code}")

soup = BeautifulSoup(response.content, "html.parser")
```

```python
# 各item要素のdescription要素のリストを得る
descriptionList = soup.findAll("description")
```

```python
# 各description要素を形態素解析
for description in descriptionList:
    # description要素の内容をプレーンテキストとして得る
    descriptionString = description.text
    print(f"Text: {descriptionString}")
    print()
    # 形態素解析
    for morpheme in tokenizer_obj.tokenize(descriptionString, tokenizer.Tokenizer.SplitMode.C):
        print(
            morpheme.surface() + "\t" +  # 表層形
            "-".join(morpheme.part_of_speech()) + "," +  # 品詞
            morpheme.dictionary_form() + "," +  # 原形
            morpheme.reading_form() + "," +  # 読み
            morpheme.normalized_form()  # 正規形
        )
    print("EOS") # 文の終端 (End of Sentence)
    print()
```
