# 遊戯王の新着商品情報

```python
# 必要なライブラリをインストール（Jupyter Notebookで実行する場合）
# !pip install beautifulsoup4 requests

# 必要なモジュールをインポート
import requests
from bs4 import BeautifulSoup

url = "https://www.yugioh-card.com/japan/products/"  # 遊戯王の商品情報ページのURL

# URLにアクセスし、ページのHTMLを取得
response = requests.get(url)

# ステータスコードが200であれば正常にアクセスできている
if response.status_code == 200:
    print("ページの取得に成功しました")
else:
    print(f"ページの取得に失敗しました: {response.status_code}")

# BeautifulSoupを使ってHTMLをパース（解析）
soup = BeautifulSoup(response.content, "html.parser")

# パースされたHTML全体を表示
print(soup.prettify())
```

```python
# 特定の要素をすべて検索（例: 'div'タグの'information modal' というクラス名）
itemList = soup.find_all("div", "information modal")

productList = []

for item in itemList:
    name = item.find("span", "subject").text
    date = item.find("dd", "date").text
    price = item.find("dd", "price").text
    link = item.find("dd", "links")
    if link:
        link = link.find("a")["href"]

    productList.append({
        "name": name,
        "date": date,
        "price": price,
        "link": link,
    })

print(productList)
```
